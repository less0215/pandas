## Scatterplot 그리기 기본

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/global_internet_users/global_internet_users.csv')

### 그래프 그릴 준비
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

### 산점도 그래프 그리기
sns.scatterplot(x='Year', y='No. of Internet Users', data=df, ax=ax)
```

## Scatterplot : hue 인자로 그룹별 색깔 지정

```python
# df.head()

### 'Entity' 고유값 확인
# df['Entity'].unique()

### 'Entity'의 고유값 중 3가지 값만 사용하기 위해 따로 변수 저장
entities = ['China', 'India', 'Finland']

###  추출한 값을 'Entity'에 저장
  # isin()은 소괄호 안에 있는 값을 True 혹은 False로 반환
  # loc[]는 대괄호 안에 있는 값 중 True에 해당하는 행만 추출
df = df.loc[df['Entity'].isin(entities)]

### 그래프 그릴 준비하기
fig, ax = plt.subplots()

### 그래프 그리기
  # hue 인자 사용해서 변수 별로 색 다르게 지정하기.
  # palette 인자는 색의 밝기 조절 해주는 인자
sns.scatterplot(
    x='Year', y='No. of Internet Users', data=df, ax=ax,
    hue='Entity', palette = 'bright'
)
```

## Scatterplot : style 인자로 그룹별 모양 지정

```python
### 그래프 그릴 준비
fig, ax = plt.subplots()

### 그래프 그리기
  # Entity의 모양을 o,x,세모로 변경
  # 사이즈(s)는 100
sns.scatterplot(
    x='Year', y='No. of Internet Users', data=df, ax=ax,
    style='Entity', markers=['o', '^', 'X'], s=100
)
```

## Scatterplot : sizes 인자로 그룹별 크기 지정

```python
### 그래프 그릴 준비
fig, ax = plt.subplots()

### 그래프 그리기
  # Entity의 모양을 o,x,세모로 변경
  # 사이즈(s)는 100
sns.scatterplot(
    x='Year', y='No. of Internet Users', data=df, ax=ax,
    size='Entity', sizes=(40, 200)
)
```

### Plotly : 산점도 그리기

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/global_internet_users/global_internet_users.csv')

### plotly express 이용해서 산점도 차트 그리기
  ### matplotlib처럼 따로 subplot을 그릴 필요가 없다.

import plotly.express as px

fig = px.scatter(
    data_frame=df, x="Year", y="No. of Internet Users",
    width = 400, height = 400
)

fig.show()
```

### Plotly : 색 지정해서 산점도 그리기

```python
### 분류 하고 싶은 것 지정하기

# df.head()
entities = ['China', 'India', 'Finland']
df = df.loc[df['Entity'].isin(entities)]

### 색 지정해서 그래프 그리기
import plotly.express as px

fig = px.scatter(
    data_frame=df, x='Year', y='No. of Internet Users',
    width=400, height=400, color='Entity'
)

fig.show()
```
