## 날짜 데이터(datetime)로 데이터 타입 변경하기

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/Uber/Uber.csv')
df.info()
```

```python
# df.head()

### "START_DATE*"의 데이터 타입을 datetime으로 변경하기
  # errors= 'coerce'는 데이터 타입은 datetime으로 강제로 변환하는걸 돕는 인자
    # 변환하려는 데이터에 결측치가 있으면 오류가 발생할 수 있기 때문에 강제로 변환
df['START_DATE*'] = pd.to_datetime(df['START_DATE*'], errors= 'coerce')
# df.info()

### "END_DATE*"의 데이터 타입을 datetime으로 변경하기
df['END_DATE*'] = pd.to_datetime(df['END_DATE*'], errors= 'coerce')

### START_DATE*, END_DATE*을 오름차 순으로 정렬하기
df = df.sort_values(['START_DATE*', 'END_DATE*'])
# df.head()
```
