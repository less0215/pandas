```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/weight_height/weight_height.csv')
df.describe()
```

```python
### scatterplot 그려서 이상치 있는지 확인하기
import seaborn as sns
# sns.scatterplot(x= 'Weight', y= 'Height', data=df)

### 이상치가 어떤 행에 있는지 확인해보기
df[df['Weight'] > 350].index
df['Weight'][2014]
```

```python
### Weight 범위를 350 이하로 제한한 상태로 기술통계 값 보기

df_new = df.query('Weight < 350')
df_new.describe()
```

```python
### Weight가 99.99%에 해당하는 값 보기
  # criteria는 criterion의 복수형.
    # '기준'이라는 뜻을 가짐. 99.99%는 '모든 데이터의 기준'이라고 해석할 수 있음.
criteria = df['Weight'].quantile(0.9999)
# criteria


### criteria보다 작은 Weight 값으로 scatterplot 그리기
df_new2 = df[df['Weight'] < criteria]

sns.scatterplot(x = 'Weight', y = 'Height', data = df_new2)
```
