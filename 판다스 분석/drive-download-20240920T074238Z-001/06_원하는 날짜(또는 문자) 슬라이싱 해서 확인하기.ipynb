## 원하는 날짜 슬라이싱 해서 확인하기

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/Covid19-US/us_confirmed.csv')
df.info()
```

```python
### 'Date'열은 인덱스로 변환한 후에 오름차 순 정렬
  # 20년 1월부터 22년 4월까지 데이터 반환
df.set_index('Date').sort_values('Date')
```

```python
### 원하는 날짜 슬라이싱 해서 확인하기

df = df.set_index('Date').sort_values('Date')

df['2020-01':'2020-02']
```

## 원하는 문자 슬라이싱 해서 확인하기

```python
df = df.reset_index().set_index('Province/State').sort_index()
df.index.unique()
# df.index.nunique()

# Df까지 슬라싱한 경우, f 이전 값인 e까지 슬라이싱 됨.
df['Ca':'Df']
```
