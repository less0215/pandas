```python
### 필요한 라이브러리 설치

# 선형회귀 분석에 필요한 라이브러리
from scipy.stats import linregress
# 그래프를 그리기 위한 라이브러리
import matplotlib.pyplot as plt
# 그래프를 이쁘게 꾸밀 수 있는 라이브러리
import seaborn as sns

### 결측치 제거
df = df.dropna()

### 선형회귀분석 준비
  # 회귀분석이란?
    # 두 변수간의 상관관계를 알아보는 행위
      # 회귀는 '돌아가다'는 뜻을 가짐. 회귀라는 이름을 가진 건 회귀분석을 제안한 통계학자 때문.
        # 통계학자는 '자녀의 키는 전체 인구의 평균 키로 되돌아간다'는 말을 하며 회귀분석을 제안해서.
  # 선형회귀란?
    # 회귀분석 종류 중 하나.

# s(slope, 기울기) , i(intercept, y절편), r(상관계수), _, _
  # y = 2.5x + 10일때 2.5가 s, 10이 i. 해당 식에서 0.8이라는 상관계수가 나왔다면 '강한 양의 상관관계를 지녔다'고 표현
    # _, _는 파이썬에서 사용하지 않을 값을 무시할 때 쓰는 형식.
      # p-value, 표준오차(stderr)가 들어가야 하는데, 해당 식에서는 필요 없어서 _로 표기
s, i, r, _, _ = linregress(df['MILES*'], df['DURATION*'])

### 그래프 그릴 준비
fig, ax = plt.subplots()

### 그래프 그리기
sns.regplot(
    x='MILES*', y='DURATION*', data=df, ax=ax,
    # line_kws는 선형회귀를 어떤 식으로 표현할지 정하는 인자
      # y는 일차식을 뜻하고, R^2는 결정계수를 뜻함.
        # 결정계수는 상관계수를 제곱한 결과. 상관계수를 제곱하면, 음수가 나오지 않음.
    # format()는 중괄호에 값을 넣는 메서드
      # 작은 따옴표 밖에서 사용한 이유는 문자열로 인식 되지 않기 위해서.
    line_kws={
        'label' : 'y={:.2f}x + {:.2f}, R^2 = {:.2f}'.format(s, i, r**2)
    }
)

### 범례 표시
  # 지도에서 기호 같은 걸 legend라고 부름.
plt.legend()
```
