## 여러 변수에서 특정 변수 추출 후 해당 변수의 데이터 확인하기

```python
### 데이터로드
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/Covid19-US/us_confirmed.csv')
# df.info()

### 데이터변환
df['Date'] = pd.to_datetime(df['Date'])
# df.info()

### Date 컬럼 인덱스로 설정 후 오름차 순 정렬

df = df.set_index('Date').sort_values('Date')
# df

### 'Province/State' unique 개수 세기
df['Province/State'].nunique()
```

```python
### 'Province/State' 변수 중 상위 3개만 추출
states = df['Province/State'].unique()[0:3]

### 추출한 값 'Province/State'에 저장
df = df[df['Province/State'].isin(states)]

### 확인
# df
df['Province/State'].nunique()
```
