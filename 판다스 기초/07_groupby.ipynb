```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
# 남녀별 생존 평균율은?

df.groupby('sex')['survived'].mean()
```

```python
'''
칼럼을 2개 이상 묶을 때는 대괄호 사용
'''


# 좌석 등급에 따른 남녀별 평균 생존율은?

df.groupby(['sex', 'class'])['survived'].mean()
```

```python
'''
데이터를 집계하는 방식이 2개 이상이면, agg() (aggregate)를 사용
'''

# 좌석 등급에 따른 남녀별 평균 생존율, 생존수는?

df.groupby(['sex', 'class'])['survived'].agg(['mean', 'count'])
```

```python
'''
agg()에서는 특정 열에 대해 어떤 식으로 데이터를 집게할지 방식을 정할 수 있다.
  2개 이상 조건을 걸 경우, {} 안에 조건을 넣는다.
'''

# 좌석 등급에 따른 남녀 생존률, 나이는?
     # (생존률은 평균값으로, 나이는 가장 많은 나이를 선택해서 출력할 것)

df.groupby(['sex', 'class'])[['survived', 'age']].agg({'survived' : 'mean', 'age' : 'max'})
```

```python
# 좌석 등급에 따른 성별·좌석별 나이의 극단값을 구하는 기준은?
  # 극단값이란, 이상치를 뜻함.
    # 해당 공식을 이용해서 '극단값을 구하는 기준'을 구함.
    #   그런 다음 해당 공식을 이용해서 튀는 값을 구한다.

import numpy as np

# 매개변수(나이)에 전달되는 값은 .apply가 적용된 age의 데이터가 전달됨.
def 극단값구하는기준(나이) :
  # _를 이름 앞에 붙인 이유는 변수명으로 숫자가 가장 먼저 올 수 없기 때문에.
  _3사분위수 = 나이.quantile(.75)
  _1사분위수 = 나이.quantile(.25)
  # 3사분위수 - 1사분위수 * 1.5는 극단값 구하는 기준을 구하는 공식.
  return (np.abs(_3사분위수 - _1사분위수) * 1.5)

df.groupby(['sex', 'class'])['age'].apply(극단값구하는기준)
```

```python
df = sns.load_dataset('penguins')
# 결측치 개수 확인
df.isna().sum()
```

```python
# 팽귄 종별로 묶고, 종별 'bill_length_mm','bill_depth_mm','flipper_length_mm' 내에 있는 결측치를 평균값으로 바꾸기

df.groupby('species')[['bill_length_mm','bill_depth_mm','flipper_length_mm']].apply(lambda x : x.fillna(x.mean()))
```

```python
import pandas as pd


df = pd.DataFrame(
  {'group':['A','A','A','B','B'],
  'value':[1, 1, 1, 10, 10]}
)
```

```python
# group별로 묶고 value값의 합 구하기

df.groupby('group')['value'].sum()
```

```python
# 그룹 내에 있는 값을 '다른 이름으로 변경'해서 value 합 구하기

df.groupby([0,0,1,1,1])['value'].sum()
```
