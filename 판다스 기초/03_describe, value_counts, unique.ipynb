### value_counts() : 범주형 데이터에 있는 정보 확인

```python
# 'Review' 칼럼에 있는 데이터 확인하기
  # null 값은 제외하고 값이 출력된다

df['Review'].value_counts()
```

```python
'''

df.loc[df['Review'] == 'Superb 9.0', 'Review'] = 'Superb '

▶ 작성순서
df에 있는 'Review' 열 내에 있는 'Superb 9.0'를 'Superb '로 바꾸겠다
  df.loc[df['Review'] == Superb 9.0] = 'Superb '
이때 결과 값은 Review 열만 보이게 하고 싶슾니다.
  df.loc[df['Review'] == Superb 9.0, 'Review'] = 'Superb '
변경한 걸 다시 저장합니다.
  df.loc[df['Review'] == 'Superb 9.0', 'Review'] = 'Superb '

'''


df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bookings/Bookings.csv')
df['Review'].value_counts()

# Review 칼럼 내에 있는 Superb 9.0를 Superb 값에 통합하기
df.loc[df['Review'] == 'Superb 9.0', 'Review'] = 'Superb '
df['Review'].value_counts()

# Review 칼럼 내에 있는 Exceptional 10를 Exceptional 값에 통합하기
df.loc[df['Review'] == 'Exceptional 10', 'Review'] = 'Exceptional '
df['Review'].value_counts()
```

### unique() : 유니크한 값만 확인하기

map(lambda) 이용해서 각 행에 있는 값 변경하기

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bookings/Bookings.csv')
df['Total_Review'].unique()

# 쉼표(,) 제거 하기
# df['Total_Review'].map(replace(',',''))
df['Total_Review'] = df['Total_Review'].map(lambda x : str(x).replace(',',''))

# review 제거하기
df['Total_Review'] = df['Total_Review'].map(lambda x : str(x).replace('review', '').strip())


# external 제거하기
df['Total_Review'] = df['Total_Review'].map(lambda x : str(x).replace('external', '').strip())

df['Total_Review'].unique()
```

astype(), describe() : 데이터타입 변경 후 기술통계 확인하기

```python
df['Total_Review'] = df['Total_Review'].astype('float')
df['Total_Review'].info()
df['Total_Review'].describe()
```
