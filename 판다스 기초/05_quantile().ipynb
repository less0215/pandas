### quantile() : 수치형 데이터 분위수 구하기

````python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bookings/Bookings.csv')

# 쉼표(,) 제거 하기
# df['Total_Review'].map(replace(',',''))
df['Total_Review'] = df['Total_Review'].map(lambda x : str(x).replace(',',''))

# review 제거하기
df['Total_Review'] = df['Total_Review'].map(lambda x : str(x).replace('review', '').strip())

# external 제거하기
df['Total_Review'] = df['Total_Review'].map(lambda x : str(x).replace('external', '').strip())

df['Total_Review'] = df['Total_Review'].astype('float')
df.info()
````

````python
quantile = [0, 0.2, 0.4, 0.6, 0.8, 1]

for idx in quantile:
  q = df['Total_Review'].quantile(idx,interpolation = 'lower')
  print(f'quantile({idx}) is {q}')
````

quantile() 예시

네, 말씀하신 코드를 바탕으로 더 쉽게 설명해드리겠습니다.

먼저, 이 코드가 하는 일을 간단히 설명하면, 'Total_Review' 컬럼의 데이터를 크기 순으로 나열하고, 특정 지점들의 값을 찾는 거예요.

`quantile = [0, 0.2, 0.4, 0.6, 0.8, 1]`는 우리가 찾고 싶은 지점들을 나타내요.
- 0은 가장 작은 값
- 0.2는 아래에서 20% 지점
- 0.4는 아래에서 40% 지점
- 0.6은 아래에서 60% 지점
- 0.8은 아래에서 80% 지점
- 1은 가장 큰 값

그리고 `interpolation = 'lower'`는 "계산된 위치보다 작거나 같은 가장 가까운 실제 데이터 값을 선택하라"는 뜻이에요.

이해를 돕기 위해 간단한 예시를 들어볼게요:

```python
import pandas as pd

# 예시 데이터: 10명 학생의 시험 점수
df = pd.DataFrame({'Total_Review': [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]})

quantile = [0, 0.2, 0.4, 0.6, 0.8, 1]

for idx in quantile:
    q = df['Total_Review'].quantile(idx, interpolation='lower')
    print(f'quantile({idx}) is {q}')
```

이 코드를 실행하면 결과는 다음과 같을 거예요:

```
quantile(0) is 55
quantile(0.2) is 60
quantile(0.4) is 70
quantile(0.6) is 80
quantile(0.8) is 90
quantile(1) is 100
```

이 결과를 해석해보면:
- 가장 낮은 점수(0%)는 55점
- 아래에서 20% 지점의 점수는 60점
- 아래에서 40% 지점의 점수는 70점
- 아래에서 60% 지점의 점수는 80점
- 아래에서 80% 지점의 점수는 90점
- 가장 높은 점수(100%)는 100점

'lower' 방식을 사용했기 때문에, 정확히 그 위치에 해당하는 값이 없으면 그보다 작은 가장 가까운 실제 점수를 선택해요.

예를 들어, 정확히 20% 지점은 61점이겠지만, 실제 데이터에 61점이 없으므로 그보다 작은 60점을 선택한 거죠.

이렇게 데이터의 분포를 대략적으로 파악할 수 있어요. 시험 점수의 전반적인 분포를 빠르게 이해하는 데 도움이 됩니다.
