```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/APPL_price/APPL_price.csv')
df.info()
```

```python
# df에 있는 Date 칼럼의 데이터 타입을 datetime으로 변경하기

df['Date'] = pd.to_datetime(df['Date'])
df.info()

# 변경한 date 칼럼을 인덱스로 지정하기
df = df.set_index('Date')
df.head()
```

```python
# 일주일 단위로 각 컬럼의 평균값을 출력하시오
  # 1. 데이터 불러오기
  # 2. 칼럼 데이터타입 datetime로 변경
  # 3. datetime으로 변경한 칼럼 set_index로 지정
  # 4. resample() 이용해서 일주일 단위 평균 값 구하기

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/APPL_price/APPL_price.csv')
# df.info()
df['Date'] = pd.to_datetime(df['Date'])
# df.info()

df = df.set_index('Date')
# df.head()

df.resample('7d').mean()
```
