### seaborn 이용해서 데이터셋 불러오기

```python
import seaborn as sns

df = sns.load_dataset('penguins')
df.head()
```

특정 열을 선택해서 조건에 맞는 행 출력하기

```python
df.query('bill_length_mm > 55')
```

```python
df[df['bill_length_mm'] > 55]
```

```python
df.loc[df['bill_length_mm'] > 55]
```

2개 이상 조건을 걸어서 특정 열 내에 있는 행 출력하기

```python
# bill_length_mm 컬럼 내에 55인 값 + species가 Gentoo인 행 출력

df.query('bill_length_mm > 55 and species == "Gentoo"')
```

외부변수를 활용해 필터링 하기

```python
'''
df.query('bill_length_mm >= @length and species == @species')

@ 이용해서 외부변수 불러오기
'''

length = 55
species = 'Gentoo'

df.query('bill_length_mm >= @length and species == @species')
```

```python
# species 칼럼에 filtering 외부 변수 적용해서 행 출력하기

filtering = ["Adelie", "Chinstrap"]

df.query('species.isin(@filtering)', engine='python')
```

contain() : 특정 열 내에 특정 문자열을 포함한 행 출력하기

```python
# island 내에 있는 oe 값만 출력하기

df.query('island.str.contains("oe")', engine='python')
```

species 내에 있는 문자열 중 e로 끝나는 행만 출력하기

```python
df.query('species.str.endswith("e")', engine='python')
```
