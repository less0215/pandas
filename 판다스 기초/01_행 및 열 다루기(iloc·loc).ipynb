## 데이터 불러오기 및 임의 결측치 삽입

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bike_rentals/bike_rentals.csv')
df.iloc[2:3] = np.nan
df.head()
```

## iloc·loc 행 및 열 선택하기

```python
'''
iloc[행, 열]
loc[행, 열]

iloc은 인덱스 순서(0부터)를 기준으로 행 및 열을 선택한다.
'''

df.iloc[2:5, 3:6]
```

```python
'''
loc은 인덱스 이름으로 행 및 열을 선택한다.
'''

df.loc[2:4, 'workingday':'temp']
```

## 조건 부여해서 특정 행 및 열 선택하기

```python
'''
df.loc[조건, 열]

# 열을 선택하지 않았기 때문에 모든 열을 반환
df.loc[(df['season'] != 1) & (df['weather'] != 2)]

'''


# “season”이 1이 아니면서 weather가 2가 아닌 모든 행들만 필터링
df.loc[(df['season'] != 1) & (df['weather'] != 2)]

# “season”이 1이 아니면서 weather가 2가 아닌 열만 필터링
df.loc[(df['season'] != 1) & (df['weather'] != 2), ['season', 'weather']]
```

```python
'''
'~'는 여집합.

not의 의미를 갖고 있다고 생각하면 된다.

소괄호로 묶을 때, ~는 소괄호 밖에 작성해 준다.
'''

df.loc[~(df['season'] == 1) & ~(df['weather'] == 2)]
```
