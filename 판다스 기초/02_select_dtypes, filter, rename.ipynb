### select_dtypes() : 특정 데이터 타입을 가진 열만 선택하기

int 타입인 열 선택하기

```python
df.select_dtypes(include='int')
```

int 타입이 아닌 열 선택하기

```python
df.select_dtypes(exclude='int')
```

### filter() : 조건에 맞는 행 또는` 열 선택

like : 특정 값이 포함된 행 또는 열 선택

```python
'''
1. 데이터 불러오기
2. filter()를 적용할 열 선택
3. df.filter(like='00:00:00', axis = 0)
      axis = 1 하면, 해당 열만 출력.
      axis = 0 하면, 해당 값이 포함된 모든 열까지 선택.
'''

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bike_rentals/bike_rentals.csv')
df = df.set_index('datetime')
df.filter(like='00:00:00', axis = 0)
```

items : 특정 값을 가진 행 또는 열 선택

```python
# 'humidity', 'windspeed' 열만 선택하기
df.filter(items=['humidity', 'windspeed'])
```

정규 표현식 조건에 맞는 행 또는 열 선택

```python
# 'in + 아무 문자 + s'가 포함된 열 선택하기

df.filter(regex='in.s')
```

### rename() : 행 또는 열 이름 변경하기

axis 인자 사용해서 컬럼명 변경하기

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bike_rentals/bike_rentals.csv')

### 방법1
# registered 컬럼명을 registered_user로 변경하기
# casual 컬럼명을 unregistered_user로 변경하기

df.rename({
    'registered' : 'registered_user',
    'casual' : 'unregistered_user'},
    axis = 1
)
```

axis 인자 사용하지 않고 컬럼명 변경하기

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_analysis_adv-main/datasets/bike_rentals/bike_rentals.csv')

### 방법2
# registered 컬럼명을 registered_user로 변경하기
# casual 컬럼명을 unregistered_user로 변경하기

df.rename(columns = {
    'registered' : 'registered_user',
    'casual' : 'unregistered_user'
})
```
